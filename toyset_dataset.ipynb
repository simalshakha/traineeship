```python

losses = []

for epoch in range(epochs):
    total_loss = 0
    
    for i in range(len(X)):
        x_i = X[i]
        y_i = y_true[i]
        
       
        y_pred = w * x_i + b
        
       
        loss = 0.5 * (y_i - y_pred) ** 2
        total_loss += loss.item()

        
        # Gradients
        dL_dypred = (y_pred - y_i)
        grad_w = dL_dypred * x_i
        grad_b = dL_dypred
        
        # Update
        w -= lr * grad_w
        b -= lr * grad_b
    
    losses.append(total_loss / len(X))
    if (epoch+1) % 100 == 0:
        print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

print("\nFinal weight:", w)
print("Final bias:", b)

# Predictions
y_pred_final = w * X + b

# Plot
plt.scatter(X, y_true, label="Data")
plt.plot(X, y_pred_final, color="red", label="Learned Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Loss curve
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Training Loss")
plt.show()
```
